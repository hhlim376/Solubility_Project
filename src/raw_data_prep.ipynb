```python
## Read Wang Data into dataframes and combine data
dfw1 = pd.read_excel("ci800406y_si_001.xls")
dfw2 = pd.read_excel("ci800406y_si_002.xls")
dfw3 = pd.read_excel("ci800406y_si_003.xls")
dfw4 = pd.read_excel("ci800406y_si_004.xls")
dfw5 = pd.read_excel("ci800406y_si_005.xls")
```

```python
dfw1
```

```python
## Clean up headers
df_list = [dfw1, dfw2, dfw3, dfw4, dfw5]
for df in df_list:
    df.columns = df.iloc[1]
    df.drop(df.index[:2], inplace=True)
    
```

```python
## Drop nulls
for df in df_list:
    df.dropna(inplace = True)
```

```python
## Reindex
for i in range(len(df_list)):
    df_list[i] = df_list[i].reset_index(drop=True)
```

```python
## Drop samples from delaney dataset
dfw1 = dfw1[dfw1['Data Source ID'] != 2]
```
